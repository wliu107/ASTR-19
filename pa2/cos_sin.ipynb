# import matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

# create array x [0,1] in 100 steps

```python
x=np.linspace(0,1,100)
```

# define functions that return sin and cos

```python
def sin_function(x):
    return np.sin(x)
def cos_function(x):
    return np.cos(x)
```

# plot

```python
f, axarr =plt.subplots(1,2)

axarr[0].plot(x,sin_function(x))
axarr[0].set_xlabel("x")
axarr[0].set_ylabel("sin(x)")
axarr[0].set_title(r"$\sin(x)$")

axarr[1].plot(x,cos_function(x))
axarr[1].set_xlabel("x")
axarr[1].set_ylabel("cos(x)")
axarr[1].set_title(r"$\cos(x)$")

f.subplots_adjust(wspace=0.4)
plt.savefig("sin_cos.pdf",format="pdf")
plt.show()
```
