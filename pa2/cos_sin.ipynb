# import matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

# create array x [0,1] in 100 steps

```python
x=np.linspace(0,1,100)
```

# define functions that return sin and cos

```python
def sin_function(x):
    return np.sin(x)
def cos_function(x):
    return np.cos(x)
```

# plot

```python
plt.figure()

plt.subplot(1,2,1)
plt.plot(x, sin_function(x))
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("sin(x)")

plt.subplot(1,2,2)
plt.plot(x, cos_function(x))
plt.xlabel("x")
plt.ylabel("cos(x)")
plt.title("cos(x)")

plt.savefig("sin_cos.pdf",format="pdf")
plt.show()
```
