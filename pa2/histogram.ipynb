# import matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

# use numpy to pull 1000 random numbers distributed uniformly between [0,1]. 

```python
random_num=np.random.uniform(0,1,1000)
```

# Histogram the random numbers into 100 bins, and plot the histogram.

```python
plt.figure()
plt.hist(random_num, bins=100)
plt.xlabel("value")
plt.ylabel("frequency")
plt.title("histogram")

plt.savefig("histogram",format="pdf")
```
