# import matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

# create an array x ranging from [0,1] in 100 steps.

```python
x=np.linspace(0,1,100)
```

# define a function to returns exp(x)

```python
def exp_function(x):
    return np.exp(x)
```

# make y=exp(x)

```python
y=exp_function(x)
```

# plot

```python
plt.figure()
plt.plot(x,y)
plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness")
plt.title("exp(x)")
plt.grid(True)
plt.savefig('exp.pdf',format='pdf')
plt.show()
```
