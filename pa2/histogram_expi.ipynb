# import matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Select your favorite distribution (other than uniform) and pull 1000 random numbers from that distribution. 

```python
random_num=np.random.exponential(scale=1.0, size=1000)
```

# Histogram the random numbers into 100 bins, and plot the histogram. 

```python
plt.figure()
plt.hist(random_num, bins=100)
plt.xlabel("value")
plt.ylabel("frequency")
plt.title("histogram")

plt.savefig("histogram_expo",format="pdf")
```
